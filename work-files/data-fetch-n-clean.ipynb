```python
# libraries for data manipultion
import pandas as pd
import numpy as np

# libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# finance libraries
import yfinance as yf
```

```python
tr_eikon_df = pd.read_csv('../data-files/tr_eikon_eod_data.csv')
tr_eikon_df.index = pd.to_datetime(tr_eikon_df['Date'])
tr_eikon_df = tr_eikon_df[['AAPL.O', 'MSFT.O', 'INTC.O', 'AMZN.O', 'GS.N']].dropna()
# I see some dates where no trading occurs so in this case it seems resonable to dropna()

tr_eikon_df
```

```python
# returns a heatmap on how linearly correlated all features are 

def corr_heatmap(df):
        df.reset_index(drop=True, inplace=True)
        graph_size = (12, 12)
        fig, ax = plt.subplots(figsize=graph_size)
        sns.heatmap(df.corr(), ax=ax, vmin=-1, vmax=1, cmap='coolwarm')

        plt.show()

```

```python
corr_heatmap(tr_eikon_df)
```
