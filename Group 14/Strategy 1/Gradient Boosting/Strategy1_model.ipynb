# Backtest

```python
# row wise profit
def single_side(row):
    if row['pred'] == 1:
        return row['Close'] - row['Open']
    else:
        return row['Open'] - row['Close']
```

```python
# row wise log return
def single_side_logret(row):
    if row['pred'] == 1:
        return np.log(row['Close'] / row['Open'])
    else:
        return np.log(row['Open'] / row['Close'])
```

```python
# inverted log return 
def inverted_single_side(row):
    if row['pred'] == 0:
        return np.log(row['Close'] / row['Open'])
    else:
        return np.log(row['Open'] / row['Close'])
```

```python
# profit with conf cutoff - not used 
def single_side_with_prob_cutoff(row):
    if row['pred'] == 1 and row['prob1'] > 0.6:
        return row['Close'] - row['Open']
    elif row['pred'] == 0 and row['prob0'] > 0.6:
        return row['Open'] - row['Close']
    else:
        return 0
```

```python
# row-wise long short
def long_short(row):
    return (row['Close'] - row['Open']) * row['prob1'] + (row['Open'] - row['Close']) * row['prob0']
```

```python
# row-wise long short log return
def long_short_logret(row):
    return (np.log(row['Close'] / row['Open'])) * row['prob1'] + (np.log(row['Open'] / row['Close'])) * row['prob0']
```

```python
# naive strategy 
def single(df):
    arr = []
    for index, row in df.iterrows():
        arr.append(single_side_logret(row))
    return np.array(arr).sum(), np.array(arr).std()
```

```python
# inverted naive strategy
def inverted_single(df):
    arr = []
    for index, row in df.iterrows():
        arr.append(inverted_single_side(row))
    return np.array(arr).sum(), np.array(arr).std()
```

```python
# Long short strategy
def LS(df):
    arr = []
    for index, row in df.iterrows():
        arr.append(long_short_logret(row))
    return np.array(arr).sum(), np.array(arr).std()
```

```python
# dollar cost averaging
def dollar_cost_averaging(df, interval, capital):
    shares = 0
    capital_used = 0
    for index, row in df.iterrows():
        if index % interval == 0:
            shares += capital / row['Open']
            capital_used += capital

    
    return capital_used, shares, shares * df.iloc[-1]['Close']

```

```python
# dca with volatility
def dca_vol(df, interval, capital):
    shares = 0
    capital_used = 0
    arr = []
    for index, row in df.iterrows():
        if index % interval == 0:
            shares += capital / row['Open']
            capital_used += capital
            arr.append(np.log(shares * row['Open'] / capital_used))
    

    
    return capital_used, shares, shares * df.iloc[-1]['Close'], np.std(arr), np.log(shares * df.iloc[-1]['Close'] / capital_used)
```

```python
# buy and hold 
def buy_and_hold(df, interval, capital):
    shares = 0
    arr = []
    for index, row in df.iterrows():
        if index == 0:
            shares += capital / row['Open']
        elif index % interval == 0:
            arr.append(np.log(shares * row['Open'] / capital))

    
    return shares, shares * df.iloc[-1]['Close'], np.std(arr), np.log(shares * df.iloc[-1]['Close'] / capital)
```
